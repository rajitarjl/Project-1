```python
#Import Modules needed for the new dataset in addition to the first
from pandas import DataFrame, Series
from pandasql import sqldf
```

```python
pysqldf = lambda q: sqldf(q, globals())
```

```python
#12. Join users to transactions, keeping all rows from transactions and only matching rows from users (left join)
#Use Pandas Merge with transactions & users for left join on UserId
transactions.merge(users, how='left', on='UserID')
```

```python
#13. Which transactions have a UserID not in users?
# Use the negation symbol ~ to extract rows which have a userid not in users
transactions[~transactions['UserID'].isin(users['UserID'])]
```

```python
#14. Join users to transactions, keeping only rows from transactions and users thatmatch via UserID (inner join)
#Use merge for inner join and specify the column with the 'on' parameter
transactions.merge(users, how='inner', on='UserID')
```

```python
#15. Join users to transactions, displaying all matching rows AND all non-matching rows(full outer join)
transactions.merge(users, how='outer', on='UserID')
```

```python
#16. Determine which sessions occurred on the same day each user registered

q = """
    SELECT
        u.*,
        s.SessionID, s.SessionDate
    FROM
        sessions s
    INNER JOIN
        users u 
    ON s.UserID = u.UserID AND s.SessionDate = u.Registered
    ORDER BY
        s.SessionID
"""
pysqldf(q)
```

```python
#18.Determine how much quantity of each product was purchased by each user
```

```python
#Use Repeat function to repeat the value=1 for each of the user id
df1 = pd.DataFrame({'key': np.repeat(1, users.shape[0]), 'UserID': users.UserID})
df1
```

```python
#Use Repeat function to repeat the value=1 for each of the product id
df2 = pd.DataFrame({'key': np.repeat(1, products.shape[0]), 'ProductID': products.ProductID})
df2
```

```python
user_products = pd.merge(df1, df2,on='key')[['UserID', 'ProductID']]
user_products
```

```python
# User Merge on the user_products and transactions with Left join combining/Grouping on UserId and ProductId to get the 
# quantity of each product was purchased by each user
pd.merge(user_products, transactions, how='left', on=['UserID', 'ProductID']).groupby(['UserID', 'ProductID']).apply(lambda x: pd.Series(dict(
    Quantity=x.Quantity.sum()
))).reset_index().fillna(0)
```

```python
#19. For each user, get each possible pair of pair transactions (TransactionID1, TransacationID2)
pd.merge(transactions, transactions, on='UserID')
```

```python
# 20 : Join each user to his/her first occuring transaction in the transactions table
#Using Merge to combine users and transactions and first() to view the first transaction
data=pd.merge(users, transactions.groupby('UserID').first().reset_index(), how='left', on='UserID')
data
```

```python
#Get the list of Columns with NAN
missing_info = list(data.columns[data.isnull().any()])
missing_info
```

```python
#For all columns in list find and display the number of values missing(NAN) in that column
for col in missing_info:
    num_missing = data[data[col].isnull() == True].shape[0]
    print('number missing for column {}: {}'.format(col, num_missing))
```

```python
#For all columns in list find and display the % of values missing(NAN) in that column
for col in missing_info:
    percent_missing = data[data[col].isnull() == True].shape[0] / data.shape[0]
    print('percent missing for column {}: {}'.format(
    col, percent_missing))
```
